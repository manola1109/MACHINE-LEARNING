# Removing the Stop Words

```python
Roos2 = pd.DataFrame({'Text':inaugural.raw('1941-Roosevelt.txt')}, index=[0])
Roos2['Text'] = Roos2['Text'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
display(Roos2.to_string())
```

```python
Kenn2 = pd.DataFrame({'Text':inaugural.raw('1961-Kennedy.txt')}, index=[0])
Kenn2['Text'] = Kenn2['Text'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
display(Kenn2.to_string())
```

```python
Nix2 = pd.DataFrame({'Text':inaugural.raw('1973-Nixon.txt')}, index=[0])
Nix2['Text'] = Nix2['Text'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
display(Nix2.to_string())
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
from collections import Counter
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
from collections import Counter
tokenizer = RegexpTokenizer(r'\w+')
Ros_no = tokenizer.tokenize(Roos)
set(w.title() for w in Ros_no if w.lower() not in stopwords.words())
word_count = Counter(w.title() for w in Ros_no if w.lower() not in stopwords.words())
dfros = word_count.most_common()
word_count.most_common()
```

```python
tokenizer = RegexpTokenizer(r'\w+')
Ken_no = tokenizer.tokenize(Kenn)
set(w.title() for w in Ken_no if w.lower() not in stopwords.words())
word_count = Counter(w.title() for w in Ken_no if w.lower() not in stopwords.words())
word_count.most_common()
```

```python
tokenizer = RegexpTokenizer(r'\w+')
Nix_no = tokenizer.tokenize(Nix)
set(w.title() for w in Nix_no if w.lower() not in stopwords.words())
word_count = Counter(w.title() for w in Nix_no if w.lower() not in stopwords.words())
word_count.most_common()
```

```python
# Word Cloud 
from wordcloud import WordCloud
```

```python
Roos2 = pd.DataFrame({'Text':inaugural.raw('1941-Roosevelt.txt')}, index=[0])
Roos2['Text'] = Roos2['Text'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
corpus = Roos2['Text'].apply(lambda x: ' '.join([z for z in x.split() if z not in stop])) 
corpus
wc_a = ' '.join(corpus)
```

```python
# Word Cloud 
from wordcloud import WordCloud
wordcloud = WordCloud(width = 4000, height = 4000, 
                background_color ='white', 
                min_font_size = 10, random_state=100).generate(wc_a) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off")
plt.xlabel('Word Cloud')
plt.tight_layout(pad = 0) 
plt.show()
```
